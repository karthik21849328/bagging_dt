```python
#importing all the required libraries 

import numpy as np
import pandas as pd 
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
import seaborn as sns 
```

```python
#loading the datset 

df=pd.read_csv("credit.csv")
df=pd.DataFrame(df)
```

```python
#displaying dataset to know about independent and dependent variables 
df.head()
```

```python
#by the above info we can notice that there are more number of non-numerical columns
# converting them into numerical type

for feature in df.columns:
    if df[feature].dtype=='object':
        df[feature]=pd.Categorical(df[feature]).codes
        
```

```python
#splitting the datset into training and testing 

train_set=df.head(700)
test_set=df.tail(300)

train_labels=train_set.pop('default')
test_labels=test_set.pop('default')
```

```python
dt_model=DecisionTreeClassifier(criterion='entropy')
```

```python
dt_model.fit(train_set,train_labels)
```

```python
dt_model.score(test_set,test_labels)
```

```python
dt_model.score(train_set,train_labels)
```

we can observe that there is 100% accuracy on the training set, overfitting case, but for the test case it is 69%

USING ENSEMBLE TECHNIQUE BAGGING 

```python
credit_labels=df.pop('default')
```

```python
from sklearn.ensemble import BaggingClassifier
bgcl=BaggingClassifier(n_estimators=50,max_samples=0.8,oob_score=True) # by default oob_score will be false 
bgcl=bgcl.fit(df,credit_labels)
print(bgcl.oob_score_)
```

Inference : By using ensemble technique our model accuracy increased from 69% to 74%
